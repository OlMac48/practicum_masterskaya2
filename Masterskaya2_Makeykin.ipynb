```python
# Обработка возможных NaN значений (если есть)
if melted_df['level_num'].isnull().any():
    print("\nОбнаружены NaN значения после извлечения чисел из 'level'. Проверьте данные.", \
          melted_df[melted_df['level_num'].isnull()])
    # Опционально: удаляем строки с NaN в 'level_num'
    melted_df = melted_df.dropna(subset=['level_num'])
    
# Преобразуем тип данных в integer
melted_df['level_num'] = melted_df['level_num'].astype(int)
grouped = melted_df.groupby(['category', 'level']).size().reset_index(name='count')

# Фильтруем категории, появляющиеся на более чем одной позиции
multi_position = grouped.groupby('category').filter(lambda x: x['level'].nunique() > 1)
grouped = melted_df.groupby(['category', 'level_num']).size().reset_index(name='count')
unique_positions = grouped[grouped['category'].isin(multi_position['category'])]

print("\nПолный список категорий, выстречающихся на разных позициях:\n", unique_positions)
```

```python
def get_all_unique_values(df_pur, unique_positions):
    results = []

    # Определяем максимальный уровень для создания столбцов
    max_level = unique_positions['level_num'].max()

    for index, row in unique_positions.iterrows():
        category = row['category']
        level_num = row['level_num']

        # Определяем названия фильтрующего столбца
        filter_column = f'category_level_{level_num}'

        # Проверка наличия фильтрующего столбца в df_pur
        if filter_column not in df_pur.columns:
            print(f"Столбец {filter_column} отсутствует в df_pur. Пропускаем строку index={index}.")
            continue

        # Фильтруем df_pur по значению в фильтрующем столбце
        filtered_df = df_pur[df_pur[filter_column] == category]

        # Инициализируем словарь для хранения результатов
        result_dict = {
            'category': category,
            'level_num': level_num,
            'count': row['count']
        }

        # Извлекаем уникальные значения из всех предшествующих уровней
        for lvl in range(1, level_num):
            target_column = f'category_level_{lvl}'
            if target_column in df_pur.columns:
                unique_vals = filtered_df[target_column].unique().tolist()
                result_dict[f'unique_level_{lvl}'] = unique_vals
            else:
                print(f"Столбец {target_column} отсутствует в df_pur. Вставляем значение None для level {lvl}.")
                result_dict[f'unique_level_{lvl}'] = None

        # Добавляем результат в список
        results.append(result_dict)

    # Преобразуем результаты в DataFrame
    unique_results_df = pd.DataFrame(results)
    return unique_results_df

# Получаем уникальные значения из всех предшествующих уровней
unique_values_df = get_all_unique_values(df_pur, unique_positions)
```

Т.к. целевой признак содержит информацию по 1 значению для каждого клиента - вначале агрегируем таблицы по клиенту, сгенерируем новые признаки, а затем объединим таблицы.

Признаки будут следующие: 
 
 Для таблицы покупок: 
 - Общее количество товара в заказах 
 - Общая сумма покупок 
 - Средняя сумма покупок 
 - Минимальная дата покупки 
 - Максимальная дата покупки 
 
Для таблицы рассылок: 
 - Количество рассылок 
 
 - Каждый тип события вынесем как отдельные колонки и посчитаем количество для клиента.
   Канал рассылки вынесем как отдельные колонки и посчитаем количество для каждого клиента
 - Максимальная и минимальная дата рассылки 
 - Частота рассылок

```python
# Создание столбца 'total_price'
df_pur['total_price'] = df_pur['quantity'] * df_pur['price']

# Группировка и агрегация
agg_functions = {
    'quantity': 'sum',
    'total_price': ['sum', 'mean'],
    'date': ['min', 'max'],
}

df_agg = df_pur.groupby('client_id').agg(agg_functions)
df_agg.columns = ['total_quantity', 'total_sum', 'average_sum', 'min_purchase_date', 'max_purchase_date']
df_agg = df_agg.reset_index()

# Средняя разница между покупками
df_pur_sorted = df_pur.sort_values(['client_id', 'date'])
df_pur_sorted['date_diff'] = df_pur_sorted.groupby('client_id')['date'].diff().dt.days
average_date_diff = df_pur_sorted.groupby('client_id')['date_diff'].mean().rename('average_date_diff')
df_agg = df_agg.merge(average_date_diff, on='client_id', how='left')
df_agg['average_date_diff'] = df_agg['average_date_diff'].fillna(0)

# Если есть дополнительные категориальные столбцы
categorical_columns = ['category_level_1', 'category_level_2', 'category_level_3', 'category_level_4', 'category_level_5']  

for col in categorical_columns:
    most_common = df_pur.groupby('client_id')[col].apply(lambda x: x.mode().iloc[0] if not x.mode().empty else None)
    df_agg = df_agg.merge(most_common, on='client_id', how='left')
```

```python
# Агрегация
df_agg_mes = df_mes.groupby('client_id').agg(
    total_messages=('message_id', 'count'),
    unique_event_types=('event', 'nunique'),
    unique_channels=('channel', 'nunique'),
    min_send_date=('date', 'min'),
    max_send_date=('date', 'max'),
    average_send_freq=('date', lambda x: x.diff().dt.days.mean())
).reset_index()

# One-Hot Encoding событий и каналов
event_dummies = pd.get_dummies(df_mes[['client_id', 'event']], prefix='event').groupby('client_id').sum()
channel_dummies = pd.get_dummies(df_mes[['client_id', 'channel']], prefix='channel').groupby('client_id').sum()

# Объединение всех признаков
df_agg_mes = df_agg_mes.merge(event_dummies, on='client_id').merge(channel_dummies, on='client_id')

# Замена NaN в средней частоте рассылок
df_agg_mes['average_send_freq'] = df_agg_mes['average_send_freq'].fillna(0)
```

## Обучение моделей

### Подготовка данных к обучению

Из пункта 1.4. выяснено наичие явного дисбаланса классов, поэтому будем делить выборки со стратификацией, чтобы в обеих выборках были 0 и 1, и указывать balanced при обучении моделей. Не будем использовать масштабирование признаков, тк для деревянных моделей, включая бустинги, масштабирование признаков необязательно, ибо деревья устойчивы к разному масштабу признаков в отличии от линейных моделей.

```python
# Определение колонки с целевым признаком, разделение данных
target_column = 'target'
X = df_all.drop(columns=[target_column])
y = df_all[target_column]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=TEST_SIZE, random_state=RANDOM_STATE, stratify=y)
```

Для решения задачи используем алгоритмы случайный лес - Random Forest и алгоритм градиентного бустинга дерева решений CatBoost

```python
# Параметры для Random Forest
param_grid_rf = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}


# Параметры для CatBoost
scale_pos_weight = len(y_train[y_train == 0]) / len(y_train[y_train == 1])
param_grid_cb = {
    'iterations': [100, 200, 300],
    'learning_rate': [0.01, 0.05, 0.1, 0.5],
    'depth': range(1, 6)
}
```

```python
scale_pos_weight
```

### Обучение моделей

```python
# Random Forest
rf = RandomForestClassifier(class_weight='balanced', random_state=RANDOM_STATE)
grid_rf = GridSearchCV(rf, param_grid_rf, cv=3, scoring='roc_auc', n_jobs=-1, verbose=1)
grid_rf.fit(X_train, y_train)
best_model_rf = grid_rf.best_estimator_
best_params_rf = grid_rf.best_params_
best_score_rf = grid_rf.best_score_
```

```python
print('Лучшая модель Random Forest и её параметры:\n\n', best_model_rf)
print ('Метрика ROC-AUC:', best_score_rf)
```

```python
# CatBoostClassifier
cb = CatBoostClassifier(scale_pos_weight=scale_pos_weight, random_state=RANDOM_STATE, verbose=False)
grid_cb = GridSearchCV(cb, param_grid_cb, cv=3, scoring='roc_auc', n_jobs=-1, verbose=1)
grid_cb.fit(X_train, y_train)
best_model_cb = grid_cb.best_estimator_
best_params_cb = grid_cb.best_params_
best_score_cb = grid_cb.best_score_
```

```python
print('Лучшая модель CatBoost и её параметры:\n\n', best_model_cb)
print ('Метрика ROC-AUC:', best_score_cb)
```
